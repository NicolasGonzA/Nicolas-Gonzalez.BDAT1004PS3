Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd

df = pd.read_csv ('/Users/nicolasgonzalez/Downloads/us-marriages-divorces-1867-2014.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

columns = df.columns.drop(['Population', 'Marriages', 'Divorces', 'Population'])

x_data = range (0, df.shape[0])
fig, ax = plt.subplots()

for column in columns:
    ax.plot(x_data, df[column])
    
    ax.set_title('Marriages and Divorces (per_capita) 1867-2014')
    plt.xlabel('Count')
    plt.ylabel('Per 1000')
    ax.legend(columns)
```
